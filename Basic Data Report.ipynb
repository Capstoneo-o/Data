```python
mydata=mydata.dropna(subset=['MATURITY'])
```

```python
numrecords = len(mydata)
```

```python
mydata['COUPON'].count() * 100 / numrecords
```

```python
def filter1(series,n=3): #3 sigma
  mean = series.mean()
  std = series.std()
  max_range = mean + n*std
  print(max_range)
```

```python
dfmax = filter1(mydata['YIELD'])
```

```python
def filter2(series,n=3): #3 sigma
  mean = series.mean()
  std = series.std()
  min_range = mean - n*std
  print(min_range)
```

```python
dfmin = filter2(mydata['YIELD'])
```

```python
mydata = mydata.drop(mydata[mydata["YIELD"] < dfmin].index | mydata[mydata["YIELD"] > dfmax].index)
```
